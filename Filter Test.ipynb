```python
import requests

requests.packages.urllib3.disable_warnings()

url = "https://172.16.17.30:4444/webconsole/APIController"
headers = {"Accept": "application/xml"}
xml_login = "<Login><Username>apiadmin</Username><Password passwordform='encrypt'>883C4BBDEF0B32DA7176B651133A52C1F215E15696EFD7D3506EB924A0EC0097</Password></Login>"
xml_action = """<Get>
                    <FirewallRule>
                        <Filter>
                            <key name="Name" criteria="like">DNAT</key>
                        </Filter>
                    </FirewallRule>
                </Get>
            """

xmldata = f"<Request>{xml_login}{xml_action}</Request>"

response = requests.post(url, headers=headers, data={"reqxml": xmldata}, timeout=30, verify=False)

print(response.text)
```

```python
import xmltodict
import json


def dict_print(data_dict):
    """Converts a dictionary to a JSON string with indentation and prints it."""
    # Convert dictionary to JSON string with indentation
    json_string = json.dumps(data_dict, indent=4)

    # Print the JSON string
    print(json_string)


# Example XML data
xml_data = """
<Request>
    <Login>
        <Username>apiadmin</Username>
        <Password passwordform='encrypt'>883C4BBDEF0B32DA7176B651133A52C1F215E15696EFD7D3506EB924A0EC0097</Password>
    </Login>
    <Get>
        <FirewallRule>
        </FirewallRule>
    </Get>
</Request>
"""

# Convert XML to dictionary
data_dict = xmltodict.parse(xml_data)

# Print the dictionary as JSON string
dict_print(data_dict)

# Convert dictionary to XML without XML declaration and root element
xml_data = xmltodict.unparse(data_dict, pretty=True, full_document=False)

# Print the XML
print(xml_data)
```
